2. **Experiment:** Rolling a fair die, **Event:** Number 5 occurs.

```python
freq = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]

relative_freq = []
for i in freq:
    outcomes = np.random.randint(1, 7, size=i)
    fives_count = np.sum(outcomes == 5)
    r=fives_count / i
    relative_freq.append(r)
    print(f"n = {i}, P(e) = {r}")
    print("-----------------------------------------")

plt.figure(figsize=(8, 6))
plt.plot(freq, relative_freq)
plt.axhline(y=1/6,color='r',linestyle='-')
plt.xlabel("Number of rolls")
plt.ylabel("Relative frequency of rolling a 5")
plt.title("Relative Frequency of Rolling a 5 vs Number of Rolls")
plt.show()
```

3. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
freq = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]

relative_freq = []
for i in freq:
    outcomes = np.random.randint(1, 7, size=i)
    greater_3 = np.sum(outcomes > 3)
    r=greater_3 / i
    relative_freq.append(r)
    print(f"n = {i}, P(e) = {r}")
    print("-----------------------------------------")

plt.figure(figsize=(8, 6))
plt.plot(freq, relative_freq)
plt.axhline(y=0.5,color='r',linestyle='-')
plt.xlabel("Number of rolls")
plt.ylabel("Relative frequency of rolling greater than 3 occurs ")
plt.title("Relative Frequency of Rolling a greater than 3 occurs vs Number of Rolls")
plt.show()
```

4. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
freq = [10, 50, 100, 500, 1000, 5000, 10000, 50000, 100000]

relative_freq = []
for i in freq:
    outcomes = np.random.randint(1, 7, size=(i,2)).sum(axis=1)
    fives_mean= (outcomes==5).mean()
    relative_freq.append(fives_mean)
    print(f"n = {i}, P(e) = {fives_mean}")
    print("-----------------------------------------")

plt.figure(figsize=(8, 6))
plt.plot(freq, relative_freq)
plt.axhline(y=0.11,color='r',linestyle='-')
plt.xlabel("Number of rolls")
plt.ylabel("Relative frequency of rolling sum of two dice equal 5")
plt.title("Relative Frequency of Rolling a sum of two dice equal 5")
plt.show()
```

# Problem 4:Random Variables Statistical Tools
Simulate Rolling a fair die $1000$ times. Then calculate the following:

- Sample mean
- Sample variance
- Sample standard deviation

Compare the results with the theoretical probability.

```python
rep = 1000
outcomes = np.random.randint(1, 7, size=rep)

sample_mean = np.mean(outcomes)
sample_variance = np.var(outcomes)
sample_std = np.std(outcomes)

# Theoretical
theoretical_mean = 3.5
theoretical_variance = 3.005636636636637
theoretical_std = np.sqrt(theoretical_variance)

print("Sample mean:", sample_mean)
print("Sample variance:", sample_variance)
print("Sample standard deviation:", sample_std)
print(theoretical_mean)
print(theoretical_variance)
print(theoretical_std)
```

# Problem 5: Correlation and Covariance

The CSV file includes a hourly/daily summary for Szeged, Hungary area, between 2006 and 2016. After inspecting the data:

- Calculate the covariance and correlation coeffecient between temperature and apparent temperature
- Calculate the covariance and correlation coeffecient between temperature and humidity

```python
df = pd.read_csv('weatherHistory.csv')
temperature = df['Temperature (C)'].to_numpy().reshape(-1, 1)
apparant_temp = df['Apparent Temperature (C)'].to_numpy().reshape(-1, 1)
humidity = df['Humidity'].to_numpy().reshape(-1, 1)
```

```python
plt.scatter(temperature, apparant_temp)
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Apparent Temperature (C)')
```

```python
plt.scatter(humidity, temperature)
plt.grid()
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
```

```python
covariance = np.cov(df['Temperature (C)'], df['Apparent Temperature (C)'])
print('Covariance: ',covariance[0][1] )
print('---------------------------------------------------')
print("np.cov():\n", covariance)
print('---------------------------------------------------')
Correlation_Cof=np.corrcoef(df['Temperature (C)'], df['Apparent Temperature (C)'])
print('Correlation Cof: ',Correlation_Cof[0][1] )
print('---------------------------------------------------')
print("np.corrcoef():\n", Correlation_Cof)
```
